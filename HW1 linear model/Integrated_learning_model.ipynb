```python
#输入数据集
import numpy as np
from scipy.io import loadmat
from matplotlib import pyplot as plt
from sklearn import ensemble
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB
from sklearn.tree import DecisionTreeClassifier, ExtraTreeClassifier
```

```python
data = loadmat('data1.1.mat')
X = data['X']
y = data['y']
X, y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, train_size=0.7, test_size=0.3)
X.shape
```

```python
# 将数据进行可视化
plt.scatter(X[:, 0], X[:, 1], c=y.flatten())
```

```python
y_train.shape, X_train.shape
```

```python
model = ensemble.AdaBoostClassifier(base_estimator=ExtraTreesClassifier(),n_estimators=53)
model.fit(X_train, y_train.ravel())
```

```python
acc = model.score(X_test, y_test)
print('test_acc:',acc)
```

```python
acc = model.score(X_test, y_test)
print('test_acc:',acc)
```

```python
def plot_predict(model, axes):
    x0s = np.linspace(axes[0], axes[1], 100)
    x1s = np.linspace(axes[2], axes[3], 100)
    x0, x1 = np.meshgrid(x0s, x1s)
    X = np.c_[x0.ravel(), x1.ravel()]
    y_pred = model.predict(X).reshape(x0.shape)
    y_decision = model.decision_function(X).reshape(x0.shape)
    plt.contour(x0, x1, y_pred, cmap=plt.cm.winter)
    plt.contour(x0, x1, y_decision, cmap=plt.cm.winter, alpha=0.2)
```

```python
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train.flatten())
plot_predict(model, [X_train[:, 0].min(), X_train[:, 0].max(), X_train[:, 1].min(), X_train[:, 0].max()])
```

```python
randomforest = ensemble.RandomForestClassifier(n_estimators=2000, max_depth=1, random_state=0)
randomforest.fit(X_train, y_train.ravel())
```

```python
acc = randomforest.score(X_test, y_test)
print('test_acc:',acc)
```

```python
randomforest = ensemble.RandomForestClassifier(n_estimators=800, max_depth=8, random_state=0, criterion='gini')
randomforest.fit(X_train, y_train.ravel())
```

```python
acc = randomforest.score(X_test, y_test)
print('test_acc:',acc)
```
