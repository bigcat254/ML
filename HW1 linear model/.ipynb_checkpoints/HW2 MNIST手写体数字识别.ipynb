```python
#导入所需模块
import torch
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import torch.optim as optim

import os
os.environ['KMP_DUPLICATE_LIB_OK'] = 'TRUE'
```

```python
#设置相关参数
epochs = 3
batch_size = 64
lr = 0.01
momentum = 0.5
log_interval = 10
```

```python
#下载数据集
train_dataset = torchvision.datasets.MNIST('test.pt', train=True, download=True,                                transform=transforms.Compose([                                transforms.ToTensor(),                                               transforms.Normalize((0.1307,),(0.3081,))
                                           ])
                                          )
test_dataset = torchvision.datasets.MNIST('training.pt', train=False, download=True,                                          transform=transforms.Compose([
transforms.ToTensor(),
transforms.Normalize((0.1307,),(0.3081,))
                                           ])
                                          )
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, num_workers=0, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=batch_size, num_workers=0, shuffle=False)
```

```python
#数据集可视化
sample, label = next(iter(train_loader))
fig, ax = plt.subplots(nrows=8,ncols=8,sharex=True,sharey=True)
ax = ax.flatten()
for i,c in enumerate(np.random.randint(0,1000,64)):
    img = sample[i].numpy().reshape(28,28)
    ax[i].imshow(img, cmap='Greys', interpolation='nearest')
ax[0].set_xticks([])
ax[0].set_yticks([])
plt.tight_layout()
plt.show()
```

```python
#模型定义
class Model(torch.nn.Module):

    def __init__(self):
        super(Model, self).__init__()
        self.conv1 = nn.Conv2d(1, 5, kernel_size=3) ## 5*26*26
        self.conv2 = nn.Conv2d(5, 10, kernel_size=4) ## 10*24*24
        self.conv2_drop = nn.Dropout2d()
        self.fc1 = nn.Linear(250, 50)
        self.fc2 = nn.Linear(50, 10)
    def forward(self, x):
        x = F.relu(F.max_pool2d(self.conv1(x), 2)) ## 5*26*26 ---> 5*13*13
        x = F.relu(F.max_pool2d(self.conv2_drop(self.conv2(x)), 2))  ## 10*10*10 ---> 10*5*5
        x = x.view(-1, 250)
        x = F.relu(self.fc1(x))
        x = F.dropout(x, training=self.training)
        x = self.fc2(x)
        return F.log_softmax(x)
```

```python
#模型初始化
model = Model()
Loss=nn.CrossEntropyLoss();
optimizer = optim.SGD(model.parameters(), lr=lr,
                      momentum=momentum)
train_losses = []
train_counter = []
test_losses = []
test_counter = [i*len(train_loader.dataset) for i in range(epochs + 1)]
```

```python
#模型训练
batch_done = 0
logs = []
for i in range(epochs):
    for data, label in train_loader:
        output = model(data)
        loss = Loss(output,label)
        loss.backward()#进行反向传播，计算梯度
        optimizer.step()#参数更新
        optimizer.zero_grad()
        batch_done += 1
        if batch_done % log_interval == 0:
            logs.append([batch_done,loss.item()])
            print('Epoch {}: {}/{} loss:{}'.format(i, (batch_done)%len(train_loader), len(train_loader), loss.item()))

```

```python
#可视化loss曲线
logs = np.array(logs)
plt.plot(logs[:,0],logs[:,1])
plt.show()
```

```python
#模型评估
model.eval()
correct = 0
for data, label in test_loader:
    output = model(data)
    _,pred = torch.max(output, dim=1)
    correct += float(torch.sum(pred == label))
print('准确率为:{} %'.format(correct/len(test_dataset)))
```
